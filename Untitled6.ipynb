```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("diabetes (2).csv")
df
```

```python
font1= {'family': 'serif', 'color': 'blue', 'size':20}
font2 = {'family': 'serif', 'color': 'darkred', 'size' :15}
plt.ylabel("Frequency", fontdict = font2)
plt.hist(df[ 'Glucose'], label="Glucose Level Histogram")
plt.xlabel("Glucose Range", fontdict = font2)
plt.legend(loc="best")
```

```python
newdf=df.groupby('Outcome')
non_diab=newdf.get_group(0)['SkinThickness'].mean()
diab=newdf.get_group(1)['SkinThickness'].mean()
plt.bar(['non_diab', 'diab'], [non_diab, diab], label="Revenue" )
plt.legend(loc="best")
```

```python
#2. Identify percentage of Diabetes and Non Diabetes patients in given dataset
newdf=df.groupby('Outcome')
non_diab=newdf.get_group(0)['Outcome'].count()
diab=newdf.get_group(1)['Outcome'].count()
tot=df['Outcome'].count()
print(non_diab,diab,tot)
array=[100*non_diab/tot,100*diab/tot]
plt.pie(array, labels=['Non diabetes','Diabetes'],autopct='%1.1ff%%')
plt.title('percentage of diabetes and non diabetes patient')
```

```python
# x axis values
x = [1,2,3,4,5,6]
# corresponding y axis values
y = [2,4,1,5,2,6]

# plotting the points
plt.plot(x, y, color='b', linestyle= 'dotted', linewidth = 3,
         marker='*', markerfacecolor='g', markersize=12)

# setting x and y axis range
plt.ylim(1,10)
plt.xlim(1,10)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('Some cool customizations!')

# function to show the plot
plt.show()
```
